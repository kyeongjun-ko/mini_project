```python
from selenium import webdriver
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas as pd
import time
import json
import requests
import MySQLdb
from sqlalchemy import create_engine
import pymysql
import os,re # 파이썬 내장모듈 os, re
from urllib.request import urlretrieve
from PIL import Image
import cv2
```

```python
# 3/3 미니프로젝트 4일차
# 이미지 크로울링 및 저장하기

driver_path = "C:\chromedriver_win32/chromedriver.exe"

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--incognito')

browser = webdriver.Chrome(driver_path, chrome_options = chrome_options)

for i in range (1009, 1200):
    url="https://선한영향력가게.com/store/view?idx="
    browser.get(url+str(i))
    
    try:
        WebDriverWait (browser, 1).until(EC.alert_is_present())
        alert= browser.switch_to.alert
        alert.accept()     
    except:
        # 403Error 방지: url접근변수 설정
        current_url = browser.current_url
        headers = {'User-Agent':'Mozilla/5.0'}
        req = Request(current_url, headers = headers)

        # http source 추출을 위한 변수설정
        html = urlopen(req)
        source = html.read().decode("utf-8")
        code = html.getcode()

        # http 출력로직
        if code == 200 :
            print("url 정상 추출 코드 : 200\n")
        else :
            print("HTTP-Error")

        # html을 dom화 하여서 객체화
        soup = BeautifulSoup(source)

        # 이미지 정보가 담겨있는 태그수
        img_url = soup.select("div.swiper-wrapper")
        stores = soup.select("div.store-desc")
    
        results=[]
        
        url1 = current_url
        url2 = url1[url1.rfind('=')+1:]
        results.append({
                "img_url":url2
            })
        
        for name in stores:
            store_name = name.select("em.underline")
            results.append({
                "img_title":store_name[0].text
            })
        
        print(results)
        eq=r'\(([^)]+)'   
        for img in img_url:
            img_src =img.select("div.swiper-slide")[0]["style"]
            img_src1= re.findall(eq,img_src)[0]
            #img_src2= re.findall(eq,img_src)[0][12:str.rfind('/')]
            results.append({
                 "img_src" : "https://xn--o39akkz01az4ip7f4xzwoa.com/"+img_src1
            })
        print(results)
        print("인덱스는 :", results[0]["img_url"])
        print("이름은 :", results[1]["img_title"])
        #print("주소는 :", results[1]["img_src"])

        if not os.path.exists("C:/stores_img"):
            os.makedirs("C:/stores_img")
        
        for result in results :
            # 저장할 이미지 파일명
            img_filename = results[0]["img_url"] + ".jpg" 
            try:
                if len(results[2]["img_src"]) >= 1 :
                    urlretrieve (results[2]["img_src"],
                                 "C:/stores_img/" + img_filename)
            except IndexError:
                path='C:\\Users\\BIT\\Desktop\\default.jpg'
                img = Image.open(path)
                img.save("C:\\stores_img\\"+img_filename)
                
                print (".", end="")
        print(results, type(results))


```
